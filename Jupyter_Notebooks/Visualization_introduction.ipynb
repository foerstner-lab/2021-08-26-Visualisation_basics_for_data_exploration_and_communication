## Visualization basics for data exploration and communication

### Loading required package

```python
from sklearn.datasets import load_diabetes
import seaborn as sns
```

### Loading and preparing the data set

[Description of the data set](https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+(Diagnostic))

```python
from sklearn.datasets import load_breast_cancer
```

```python
bc_bunch = load_breast_cancer(as_frame=True)
```

```python
bc_bunch.data["type"] = ["malignant" if x == 0 else "benign" for x in bc_bunch.target]
```

```python
bc = bc_bunch.data
```

```python
bc
```

```python
sns.relplot(data=bc, 
            x="mean area", 
            y="mean concavity")
```

```python
sns.relplot(data=bc, 
            x="mean area", 
            y="mean concavity", 
            size="worst texture")
```

```python
sns.relplot(data=bc, 
            x="mean area", 
            y="mean concavity", 
            size="worst texture", 
            style="type")
```

```python
sns.relplot(data=bc, 
            x="mean area", 
            y="mean concavity", 
            size="worst texture", 
            style="type",
            hue="type")
```

```python
sns.relplot(data=bc, 
            x="mean area", 
            y="mean perimeter",
            kind="line")
```
